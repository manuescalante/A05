```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
import scipy as sp
from scipy.integrate import solve_ivp, odeint
```

1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
def f1(x,t):
    return x/(t+1)

tspan = [0, 50]
x0 = np.array([0.2])
X = solve_ivp(f1,tspan,x0)
```

```python
plt.plot(X.t,X.y[0,:])
plt.grid()
```
